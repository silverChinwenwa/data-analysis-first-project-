```python
missing_director = df1[df1['director'].isna()]
print(f"Number of movies missing a director: {len(missing_director)}")
print(missing_director[['title']])
```

```python
df = df1.dropna(subset=['director'])
print(f"Remaining rows in dataset: {len(df)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

sns.scatterplot(data=df, x='release_year', y='rating', alpha=0.5, hue='rating', legend=False)

plt.title('Netflix Movies: Release Year vs. Rating', fontsize=14)
plt.xlabel('Year Movie Was Made', fontsize=12)
plt.ylabel('Content Rating', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
